```python
df = pd.read_csv("Luxury_Housing_Bangalore.csv")
df['Ticket_Price_Cr'] = pd.to_numeric(df['Ticket_Price_Cr'], errors='coerce')
plt.hist(df['Ticket_Price_Cr'].dropna(), bins=30)
plt.title("Luxury House Price Distribution")
plt.xlabel("Price (crores)")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# from chart:
# observation
# most houses are in lower price range (left side)
# few houses are very expensive (right side tail)

#this is called Right-skewed Distribution

# explaination:"the distrubution of luxzury house prices is postively skewed,indicating
# the most properties are concentrated in lower price ranges,with a few high-value luxury properties."
```

```python
plt.scatter(df['Unit_Size_Sqft'],df['Ticket_Price_Cr'])
plt.xlabel("size(Sqft)")
plt.ylabel("price(Cr)")
plt.title("size vs Price")
plt.show()
```

```python
# Buyer type analysis
```

```python
df['Buyer_Type'].value_counts().plot(kind='bar')
plt.title("buyer type distribution")
plt.show()
```
